```python
import pandas as pd
from collections import *
import re
import os
```

```python
df = pd.read_csv("../data/meta/University-Communications-clean.csv")
print df.shape
```

```python
df.columns
```

```python
df_rel = df[['Object Unique ID', 'Begin date']]
df_rel.head(20)
```

```python
filenames = ["../data/Fulltext/University Communications/" + f.split("20775/")[-1] + '-2.pdf.txt' for f in df['Object Unique ID'].tolist()]
```

```python
print 'We have', len(filenames), 'files'
filenames[:4]
```

```python
src_path, file_type = "../data/Fulltext/University Communications/", ".txt"
 
file_paths = []  
for root, dirs, files in os.walk(src_path):
    file_paths.extend([os.path.join(root, f) for f in files if f.endswith(file_type)])
```

```python
for i, filename in enumerate(filenames):
    if filename not in file_paths:
        new_filename = filename.split("-2.pdf.txt")[0] + "-1.xml.txt"
        filenames[i] = new_filename
```

```python
docs = []
for filename in filenames:
    txt = open(filename)
    docs.append(txt.read())
```

```python
docs[1].lower().count(" he ")
```

```python
Counter(re.findall(r" he [\w']+", docs[6].lower()))
```

```python
Counter(re.findall(r" she ", docs[0].lower()))
```

```python
he_counts = []
she_counts = []
word_counts = []
for doc in docs:
    he_counts.append(doc.lower().count(" he "))
    she_counts.append(doc.lower().count(" she "))
    word_counts.append(len(re.findall("[a-zA-Z_]+", doc)))
```

```python
 #Series(np.random.randn(sLength), index=df1.index)
df_rel["he_counts"] = he_counts
df_rel["she_counts"] = she_counts
df_rel["word_counts"] = word_counts
df_rel.head(50)
```

```python
df_rel.to_csv("../results/uni_communications_pronoun_counts.csv", index = False)
```
